```python
import pickle
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix
FIGURESIZE=(20,15)
FONTSIZE=18
plt.rcParams['figure.figsize'] = FIGURESIZE
plt.rcParams['font.size'] = FONTSIZE
plt.rcParams['xtick.labelsize'] = FONTSIZE
plt.rcParams['ytick.labelsize'] = FONTSIZE
```

#### Figure II - Validation set Accuracy vs Epochs

```python
test_set_evals =[]
for i in range(5):
    data = pickle.load(open(os.path.join('results_model_api/results_fold_%d.pkl'%i),'rb'))
    plt.plot(data['history']['val_sparse_categorical_accuracy'], linestyle='solid',label='rotation_%d'%i)
    test_set_evals.append(accuracy_score(np.reshape(data['outs_test'],-1), np.reshape(data['labels_test'],-1)))

        
plt.ylabel('Validation Set Accuracy')
plt.xlabel('Epochs')
plt.title('Validation Set Accuracy vs Epochs')
plt.legend()
plt.savefig('Figure_II.jpg',bbox_inches='tight', dpi=150)
```

```python
for i in range(5):
    data = pickle.load(open(os.path.join('results_model_api/results_fold_%d.pkl'%i),'rb'))
    plt.figure('fold_%d'%i)
    cm = confusion_matrix(np.reshape(data['outs_test'],-1), np.reshape(data['labels_test'],-1))
    cmn = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    sns.heatmap(cmn,annot=True, xticklabels=list(range(1,7)),yticklabels=list(range(1,7)), cmap='magma_r')
    plt.xlabel('Predicted Class')
    plt.ylabel('True Class')
    plt.title('Confusion Matrix for Fold %d'%i)
    plt.savefig('confusion_matrix_fold_%d.png'%i,bbox_inches='tight', dpi=150)
```

## Figure VIII - Histogram of Test Accuracy

```python
import numpy as np

plt.hist(test_set_evals,alpha=0.65, bins=5)

plt.axvline(np.mean(test_set_evals), color='k', linestyle='dashed', linewidth=1)

min_ylim, max_ylim = plt.ylim()
plt.text(np.mean(test_set_evals), max_ylim, 'Mean Test Accuracy: {:.2f}'.format(np.mean(test_set_evals)))
plt.xlabel('Test Set Accuracy')
plt.savefig('Figure_VIII.jpg',bbox_inches='tight', dpi=150)
```
